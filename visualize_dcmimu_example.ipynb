```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 12, 12

log = []
freq = None

with open('/Users/r/stuff/dcm-imu/c/DCM_IMU_uc/rust_orig.csv', 'r', encoding='latin-1') as f:
    for line in f:
        parts = line.split(',')
        try:
            dcmy = 0#float(parts[0])
            dcmp = 0#float(parts[1])
            dcmr = 0#float(parts[2])            
            my = float(parts[0])
            mp = float(parts[1])
            mr = float(parts[2])
            ry = float(parts[3])
            rp = float(parts[4])
            rr = float(parts[5])
            log.append((dcmy, dcmp, dcmr, my, mp, mr, ry, rp, rr))
        except Exception as e:
            print('lol, err: {}'.format(e))

# log = log[125000:150000]
dyaws = list(map(lambda x: x[0], log))
dpitches = list(map(lambda x: x[1], log))
drolls = list(map(lambda x: x[2], log))

myaws = list(map(lambda x: x[3], log))
mpitches = list(map(lambda x: x[4], log))
mrolls = list(map(lambda x: x[5], log))

ryaws = list(map(lambda x: x[6], log))
rpitches = list(map(lambda x: x[7], log))
rrolls = list(map(lambda x: x[8], log))

f, axarr = plt.subplots(3, sharex=True)

#axarr[0].plot(np.degrees(dyaws), c='c')
axarr[0].plot(np.degrees(myaws), c='m')
axarr[0].plot(np.degrees(ryaws), c='b')

#axarr[1].plot(np.degrees(dpitches), c='c')
axarr[1].plot(np.degrees(mpitches), c='m')
axarr[1].plot(np.degrees(rpitches), c='b')

# axarr[2].plot(np.degrees(drolls), c='c')
axarr[2].plot(np.degrees(mrolls), c='m')
axarr[2].plot(np.degrees(rrolls), c='b')
```
